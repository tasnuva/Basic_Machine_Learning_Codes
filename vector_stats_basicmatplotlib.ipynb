```python
#  Define the following two vectors using functions in the NumPy module.
```

```python
import numpy as np

x = np.array([1,2,3])
y = np.array([4,5,6])

print("Vector x: ",x)
print("\nVector y: ",y)
```

```python
# Adding the two vectors
```

```python
addition = x + y
print(addition)
```

```python
# Multiplying the vector x by the scalar 2
```

```python
mul_1 = x * 2
print(mul_1)
```

```python
# Multiplying the vector x to the vector y component-wise and printing the result
```

```python
mul_2 = x*y
print(mul_2)
```

```python
# Calculating the inner product of two vectors (dot product)
```

```python
inner_product = np.dot(x,y)
print(inner_product)
```

```python
#  Define the following two matrix (3 by 3) using functions in the NumPy module.
```

```python
M = np.array([(1, 2, 3), (2, 1, 3), (3, 2, 1)])
N = np.array([(4, 5, 6), (5, 4, 6), (6, 5, 4)])

print("Matrix M: ",M)
print("\nMatrix N: ",N)

```

```python
# Multiplying the matrix M and N using NumPy functions, and printing the result.
```

```python
component_mul = np.multiply(M,N)
print("Component based multiplication of M and N :")
print(component_mul)

dot_mul = np.dot(M,N)
print("\n Dot Product based multiplication M and N :")
print(dot_mul)
```

```python
# Multiplying the matrix N and M using NumPy functions, and printing the result.
```

```python
component_mul1 = np.multiply(N,M)
print("Component based multiplication of N and M :")
print(component_mul1)

dot_mul1 = np.dot(N,M)
print("\n Dot Product based multiplication of N and M :")
print(dot_mul1)
```

```python
# Finding the inverse of the matrix M using NumPy functions, and printing the result.
```

```python
M_inv =  np.linalg.inv(M)

print("Inverse Matrix of M :",M_inv)
```

```python
# Finding the inverse of the matrix N using NumPy functions, and printing the result.
```

```python
N_inv =  np.linalg.inv(N)

print("Inverse Matrix of N :",N_inv)
```

```python
# Multiplying M inverse by M, and printing the result.
```

```python
M_iden = M.dot(M_inv)

print("Identity Matrix of M:",M_iden.astype(int))
```

```python
# Multiplying N inverse by N, and printing the result.
```

```python
N_iden = N.dot(N_inv)

print("Identity Matrix of N:",N_iden.astype(int))
```

```python
# Multiply the product of M and N,  and then by the vector x, and printing the result. (hint: may need 'transpose' function to keep dimension consistent).
```

```python
M = np.array([(1, 2, 3), (2, 1, 3), (3, 2, 1)])
N = np.array([(4, 5, 6), (5, 4, 6), (6, 5, 4)])

product_mul = np.dot(M,N)

print("Shape of dot product: ",product_mul.shape)

x = np.array([1,2,3])

print("Shape of  Vector x: ",x.shape)

trans_x = x.transpose()

print("Transpose X:",trans_x)

print("Shape of transpose x: ",trans_x.shape)

mul_3 = np.dot(product_mul,trans_x)

print("Product: ",mul_3)

print("Shape: ",mul_3.shape)
```

```python
# Loading the California housing price dataset from sklearn and store data in the Pandas
```

```python
import pandas as pd 
import numpy as np
from sklearn.datasets import fetch_california_housing
housing  = fetch_california_housing()
housing_pd = pd.DataFrame(data= np.c_[housing['data'], housing['target']], columns= housing['feature_names'] + ['housing_value'])
housing_pd.head()
```

```python
# Calculate the mean and median of the columns of  'MedInc' and 'housing_value.
```

```python
MedInc_mean = housing_pd['MedInc'].mean()

print("Mean of MedInc: ",MedInc_mean)

housingval_mean = housing_pd['housing_value'].mean()

print("Mean of Housing Value: ",housingval_mean)
```

```python
# Calculate the quartiles (25th, 50th, 75th) of the columns of 'MedInc' and 'housing_value'.
```

```python
MedInc_quartiles = housing_pd.MedInc.quantile([0.25,0.5,0.75])

print("MedInc quartiles:\n",MedInc_quartiles)

housingval_quartiles = housing_pd.housing_value.quantile([0.25,0.5,0.75])

print("\nHousing Value quartiles:\n",housingval_quartiles)
```
