# Perceptron Model Best Weight Update Prediction

## The purpose of this project is to predict best weight update for a feedforward neural network model 

## tool used for visualization : matplotlib

## model used for prediction:  Perception Train and Test method

## tools used for data validation: confusion_matrix, precision_score, recall_score,f1_score



```python
import numpy as np
```

```python
X_train = np.array([[8,8],[0,1],[6,4],[3,3],[4,8],[1,1],[4,6],[2,5]])
y_train = np.array([1,-1,+1,-1,+1,-1,+1,+1])
```

```python
# Implementing Perceptron
```

```python
def Perceptron_Train(X,y,max_iter=100,verbose=0):
  weights = np.array([1,1])
  bias = 0
  accuracy_log = []
  for epoch in range(0,max_iter):
    if verbose: print("Start epoch:",epoch)
    errors = 0
    accuracy = 0
    for n in range(0,X.shape[0]):
      if verbose: print('Start Data point:',n)
      print('Data point:',X[n])
      feature = X[n]
      y_label = y[n]
      output = np.dot(weights,X[n])+bias
      print(y_label*output)
      if y_label * output <=0:
        if verbose: print("Start update weights: ")
        weights = weights + y_label * feature
        bias = bias + y_label
        errors += 1
      else:
        if verbose: print("Do nothing")
      if verbose: print("Latest weight:",weights,"bias: ",bias,end="\n\n")
      if verbose: print("Total errors after epoch:",epoch,": ",errors)
    accuracy = 1 - errors/X.shape[0]
    accuracy_log.append(accuracy)
    if errors == 0:
      print("Found classifier with errors: ",errors)
      print("Weights: ",weights)
      print("bias: ",bias)
      break
  return weights,bias,accuracy_log





```

```python
def Perceptron_Test(w,b,X):
  output_label = []
  for n in range(0,X.shape[0]):
    feature = X[n]
    output = np.dot(w,X[n])+b
    if output > 0:
      output_label.append(1)
    else:
      output_label.append(-1)
  return output_label
```

```python
best_weights,bias,accuracy_log=Perceptron_Train(X_train,y_train,max_iter=1,verbose=1)
```

```python
best_weights,bias,accuracy_log=Perceptron_Train(X_train,y_train,max_iter=100,verbose=1)
```

```python
# Visualizing linear boundary
```

```python
import matplotlib.pyplot as plt
```

```python
 slope =  (-(bias/best_weights[1])/(bias/best_weights[0]))
 intercept = -bias/best_weights[1]
```

```python
def linearboundary(inputs,labels,slope, intercept):
    plt.figure(figsize=(10,6))
    plt.grid(True)
    column = []
    for i in range(0,len(labels)):
      if y_train[i] == -1:
        column.append('+1')
      else:
        column.append('-1')
    
    for input,label in zip(inputs,labels):
        plt.plot(input[0],input[1],'ro' if (labels[label] == -1) else 'ko', label=column[label])
        column[label] = "_nolegend_"
    axes = plt.gca()
    x_vals = X_train[:, 0]
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
    plt.legend()
```

```python
linearboundary(X_train,y_train,slope,intercept)
```

```python
# Finding test accuracy on the test data
```

```python
X_test = np.array([[8,1],[0,2],[6,5],[3,2],[4,2],[1,1],[4,7],[2,6]])
y_test = np.array([1,-1,+1,-1,+1,-1,+1,+1])
```

```python
best_weights_test,bias_test,accuracy_log=Perceptron_Train(X_test,y_test,max_iter=100,verbose=1)
```

```python
predicted_labels = Perceptron_Test(best_weights_test,bias_test,X_test)
print(predicted_labels)
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score
print("Accuracy Score for Test Set: ",metrics.accuracy_score(y_test, predicted_labels))
```

```python
from sklearn.metrics import classification_report
print("Classification report for Test Set: \n",metrics.classification_report(y_test, predicted_labels))
```
